```python
import pandas as pd
import sqlite3

# Load CSV file into DataFrame
df = pd.read_csv("Healthcare_Claims_Dataset_300.csv")

# Create in-memory SQLite DB
conn = sqlite3.connect(":memory:")

# Load DataFrame into SQL table
df.to_sql("Claims", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT Procedure_Name, SUM(Claim_Cost) AS Total_Cost
FROM Claims
GROUP BY Procedure_Name
ORDER BY Total_Cost DESC
LIMIT 10;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# @title Procedure_Name

from matplotlib import pyplot as plt
import seaborn as sns
result.groupby('Procedure_Name').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
query = """
SELECT Procedure_Name, SUM(Claim_Cost) AS Total_Cost
FROM Claims
GROUP BY Procedure_Name
ORDER BY Total_Cost DESC
LIMIT 10;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# @title Procedure_Name vs Total_Cost

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(result['Procedure_Name'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(result, x='Total_Cost', y='Procedure_Name', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
query = """
SELECT Provider_Name, SUM(Claim_Cost) AS Total_Cost
FROM Claims
GROUP BY Provider_Name
ORDER BY Total_Cost DESC
LIMIT 5;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# @title Total_Cost

from matplotlib import pyplot as plt
result['Total_Cost'].plot(kind='line', figsize=(8, 4), title='Total_Cost')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
query = """
SELECT Procedure_Name, COUNT(*) AS Frequency
FROM Claims
GROUP BY Procedure_Name
ORDER BY Frequency DESC;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# @title Frequency

from matplotlib import pyplot as plt
result['Frequency'].plot(kind='hist', bins=20, title='Frequency')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
query = """
SELECT Procedure_Name, Provider_Name, SUM(Claim_Cost) AS Total_Cost
FROM Claims
GROUP BY Procedure_Name, Provider_Name
ORDER BY Procedure_Name, Total_Cost DESC;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# @title Procedure_Name vs Total_Cost

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(result['Procedure_Name'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(result, x='Total_Cost', y='Procedure_Name', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
